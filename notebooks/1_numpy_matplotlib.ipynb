### b) Visualisere innholdet i Medals.xlsx

I denne oppgaven skal du visualisere innholdet i filen ved hjelp av Python-biblioteker som Matplotlib.
Bruk Matplotlib til å lage grafer som viser fordelingen av medaljer.    


```python
# Skriv koden din her
medaljer = df[["Team/NOC", 'Gold', 'Silver', 'Bronze']].set_index('Team/NOC')

medaljer.plot(kind='barh', stacked=True, figsize=(10, 20), color=['gold', 'silver', 'brown'])

plt.title('Fordeling av medaljer per land', fontsize=16)
plt.xlabel('Land', fontsize=12)
plt.ylabel('Antall medaljer', fontsize=12)
plt.legend(title='Type medalje', fontsize=10)
plt.xticks(rotation=45, fontsize=10)
plt.tight_layout()

plt.show()
```
